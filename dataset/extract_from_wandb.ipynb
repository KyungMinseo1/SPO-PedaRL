```python
import os, json, shutil
from tqdm import tqdm

os.environ["WANDB_SILENT"] = "true"

api = wandb.Api(api_key="", overrides={"silent": True})
all_data = []

for i in tqdm(range(1, 207)):
    try:
        artifact = api.artifact(
            f"pegoo-sungkyunkwan-university/train-rl-SPO_LAB-server/run-kxtc6ysy-nodes_batch_{i}:latest"
        )
        data_dir = artifact.download(root=f"./Wandb/batch_{i}/")
        for f in os.listdir(data_dir):
            with open(os.path.join(data_dir, f)) as fp:
                data = json.load(fp)
                if isinstance(data, list):
                    all_data.extend(data)
                else:
                    all_data.append(data)
        shutil.rmtree(f"./Wandb/batch_{i}/")
    except Exception as e:
        print(f"Error fetching batch {i}: {e}")
        continue

print(f"총 {len(all_data)}개 데이터 로드 완료")
```

```python
all_data_path = "./Wandb/all_data.json"
with open(all_data_path, "w") as f:
    json.dump(all_data, f, ensure_ascii=False, indent=2)
print(f"✅ All data saved to: {all_data_path}")

import pandas as pd
df = pd.DataFrame(all_data)
df.to_csv("./Wandb/all_data.csv", index=False)
print(f"✅ All data saved to: ./Wandb/all_data.csv")
```

```python
import pandas as pd
import os, json, shutil
from tqdm import tqdm

with open("./dataset/all_conversations.json") as f:
    conversations = json.load(f)
df = pd.DataFrame(conversations)
df.head()
```

```python
columns = df["columns"][0]
data_list = df.data
data = [d for data in data_list for d in data if len(d) == len(columns)]
n_df = pd.DataFrame(data=data, columns = columns)
n_df.head()
```

```python
n_df.to_json("./dataset/all_conversations_refined.json", orient="records", force_ascii=False, indent=2)
```

```python
len(n_df.Problem.unique())
```

```python
n_df = n_df.sort_values(by="Problem").reset_index(drop=True)
```

```python
n_df.head()
```

```python
import ast

def turn_split(conversation_str):
    try:
        conversation_list = ast.literal_eval(conversation_str)
        split_list = []
        for idx, turn in enumerate(conversation_list):
            if turn["role"] == 'teacher':
                split_list.append(conversation_list[:idx])
        return split_list
    except Exception as e:
        print(f"Error parsing conversation: {e}")
        return None

n_df["Turn_Split"] = n_df["Conversation"].apply(turn_split)
```

```python
n_df.to_json("./dataset/all_conversations_turn_segmented.json", orient="records", force_ascii=False, indent=2)
```

```python
import os, json, shutil
from tqdm import tqdm

os.environ["WANDB_SILENT"] = "true"

api = wandb.Api(api_key="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", overrides={"silent": True})
all_data = []

for i in tqdm(range(1, 207)):
    try:
        artifact = api.artifact(
            f"pegoo-sungkyunkwan-university/train-rl-SPO_LAB-server/run-kxtc6ysy-nodes_batch_{i}:latest"
        )
        data_dir = artifact.download(root=f"./Wandb/batch_{i}/")
        for f in os.listdir(data_dir):
            with open(os.path.join(data_dir, f)) as fp:
                data = json.load(fp)
                if isinstance(data, list):
                    all_data.extend(data)
                else:
                    all_data.append(data)
        shutil.rmtree(f"./Wandb/batch_{i}/")
    except Exception as e:
        print(f"Error fetching batch {i}: {e}")
        continue

print(f"총 {len(all_data)}개 데이터 로드 완료")
```

```python
all_data_path = "./Wandb/all_data.json"
with open(all_data_path, "w") as f:
    json.dump(all_data, f, ensure_ascii=False, indent=2)
print(f"✅ All data saved to: {all_data_path}")

import pandas as pd
df = pd.DataFrame(all_data)
df.to_csv("./Wandb/all_data.csv", index=False)
print(f"✅ All data saved to: ./Wandb/all_data.csv")
```
